```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import bs4
import requests
from bs4 import BeautifulSoup
from PIL import Image
import io
import re
```

```python
r = requests.get('http://www.dota2.com/heroes/')
```

```python
s = BeautifulSoup(r.text, 'lxml')
```

```python
sc = s.find_all('div',{'class':'heroIcons'})
sc
```

```python
member = sc[0].find_all('a')
member
```

```python
rows= []
for e in member:
    url_img = e.find('img')['src']
    img_name = e.find('img')['id']
    img_html = f'<img src="{url_img}" />'
    link_url =  e['href']
    href_html = f'<a href ="{link_url}">{img_name}</a>'
    rows.append((img_name,link_url,url_img,img_html,href_html))
```

```python
df = pd.DataFrame(rows,columns=['name','link','url_img','img_html','link_html'])
```

```python
df
```

```python
from IPython.display import HTML
```

```python
HTML(df[['img_html','link_html']].to_html(escape=False))
```
